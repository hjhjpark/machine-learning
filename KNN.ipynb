```python

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

N = 100
w1 = 0.5
w0 = 2
x = np.random.normal(0, 15, N).reshape(-1,1)
y = w1*x + w0 + 5*np.random.normal(0, 1, N).reshape(-1,1)

plt.figure(figsize = (10, 8))
plt.title('Data Set', fontsize = 15)
plt.plot(x, y, '.', label = 'Data')
plt.xlabel('X', fontsize = 15)
plt.ylabel('Y', fontsize = 15)
plt.legend(fontsize = 15)
plt.axis('equal')
plt.axis([-40, 40, -30, 30])
plt.grid(alpha = 0.3)
plt.show()
```

```python
from sklearn import neighbors

reg=neighbors.KNeighborsRegressor(n_neighbors=21)
reg.fit(x,y)
```

```python
print(x.shape)
xp=np.linspace(-40,40,500)
yp=reg.predict(xp.reshape(-1,1))
print(xp.shape)
print(yp.shape)
```

```python
plt.figure(figsize = (10, 8))
plt.title('Data Set', fontsize = 15)
plt.plot(x, y, '.', label = 'Data')
plt.plot(xp,yp)
plt.xlabel('X', fontsize = 15)
plt.ylabel('Y', fontsize = 15)
plt.legend(fontsize = 15)
plt.axis('equal')
plt.axis([-40, 40, -30, 30])
plt.grid(alpha = 0.3)
plt.show()
```

```python

m = 1000
X = -1.5 + 3*np.random.uniform(size = (m,2))

y = np.zeros([m,1])
for i in range(m):
    if np.linalg.norm(X[i,:], 2) <= 1:
        if np.random.uniform() < 0.05:
            y[i] = 0
        else:
            y[i] = 1    
    else:
        if np.random.uniform() < 0.05:
            y[i] = 1
        else:
            y[i] = 0
            
C1 = np.where(y == 1)[0]
C0 = np.where(y == 0)[0]
```

```python
clf=neighbors.KNeighborsClassifier(n_neighbors=3)
clf.fit(X,y)
```

```python
pred=clf.predict(xp)
pred_cls=np.where(pred==1)[0]
print(pred_cls.shape)
```

```python
plt.figure(figsize=(8,8))
plt.plot(X[C1,0],X[C1,1],'o',label='C1')
plt.plot(X[C0,0],X[C0,1],'o',label='C0')
plt.plot(xp[pred_cls,0],xp[pred_cls,1],'s',color='red',alpha=0.5,markersize=1)
plt.show()
plt.close()
```
