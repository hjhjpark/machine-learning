```python

A = np.array([[11.08, 6.82, 1.76, -6.82],
              [2.50, -1.01, -2.60, 1.19],
              [-4.88, -5.07, -3.21, 5.20],
              [-0.49, 1.52, 2.07, -1.66],
              [-14.04, -12.40, -6.66, 12.65],
              [0.27, -8.51, -10.19, 9.15],
              [9.53, -9.84, -17.00, 11.00],
              [-12.01, 3.64, 11.10, -4.48]])
A=np.asmatrix(A)
```

```python
U,S,VT=np.linalg.svd(A,full_matrices=True)
U=np.asmatrix(U)
S=np.asmatrix(np.vstack([np.diag(S),np.zeros((4,4))]))
VT=np.asmatrix(VT)
print(U.shape)
print(S.shape)
print(VT.shape)
```

```python
print(A)
print(U*S*VT)
```

```python

from scipy import io

faces = io.loadmat('./faces1.mat')
face_mat = faces['faces']
```

```python
print(face_mat.shape)
```

```python
mat=face_mat.reshape(-1,12)
mat=np.asmatrix(mat)
print(mat.shape)
```

```python
S=(1/mat.shape[0])*mat.T*mat
print(S.shape)
```

```python
U,S,VT=np.linalg.svd(mat,full_matrices=False)
U=np.asmatrix(U)
s=S.copy()
S=np.asmatrix(np.diag(S))
VT=np.asmatrix(VT)
print(U.shape)
print(S.shape)
print(VT.shape)
```

```python
plt.stem(np.sqrt(s))
plt.show()
```

```python
for i in range(S.shape[0]):
    plt.imshow(U[:,i].reshape(250,200),'gray')
    plt.show()
```
