```python
from sklearn.decomposition import FastICA
import matplotlib.pyplot as plt
import numpy as np
```

```python

S1 = plt.imread('./Suhin.jpg')
S2 = plt.imread('./nature.jpg')
```

```python
print(S1.shape)
print(S2.shape)
```

```python
A = np.matrix([[3/4, 1/5],
              [1/2, 2/3]])
```

```python
x1=A[0,0]*S1+A[0,1]*S2
x2=A[1,0]*S1+A[1,1]*S2
```

```python
plt.imshow(np.uint8(x1))
plt.show()
```

```python
plt.imshow(np.uint8(x2))
plt.show()
```

```python
x1=x1.reshape(-1,1)
x2=x2.reshape(-1,1)
mix=np.hstack([x1,x2])
ica=FastICA(n_components=2)
ica_img=ica.fit_transform(mix)
```

```python
print(ica_img.shape)
```

```python
from sklearn.preprocessing import minmax_scale

plt.imshow(np.uint8(minmax_scale(ica_img[:,0])*255).reshape(400,250,3))
plt.show()
```

```python
import cv2
img = cv2.cvtColor(np.uint8(minmax_scale(ica_img[:,1])*255).reshape(400,250,3), cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```
